```python
import os
import json
import pandas as pd
```

```python
# Set the path to your Spotify folder
folder_path = os.path.expanduser("~/Desktop/Spotify")
```

```python
folder_path
```

```python
# Get all JSON files in the folder that match the naming pattern
json_files = sorted([
    f for f in os.listdir(folder_path)
    if f.startswith("Streaming_History_Audio_") and f.endswith(".json")
])
```

```python
# Read and combine all JSON files
all_data = []
for file_name in json_files:
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        all_data.extend(data)
```

```python
# Convert to DataFrame
df = pd.DataFrame(all_data)
```

```python
vpn_corrections = {
    'SG': 'IN',  # Singapore ➝ India
    'FR': 'US',  # France ➝ United States
    'NL': 'US',  # Netherlands ➝ United States
    'GB': 'US',  # United Kingdom ➝ United States
    'JP': 'US'   # Japan ➝ United States
}

# Apply corrections to 'conn_country'
df['conn_country'] = df['conn_country'].replace(vpn_corrections)

# Re-map corrected 'conn_country' to full names
import pycountry
country_mapping = {country.alpha_2: country.name for country in pycountry.countries}
df['conn_country_full'] = df['conn_country'].map(country_mapping)
```

```python
import pandas as pd
import pytz

# Ensure ts is parsed as timezone-aware UTC
df['ts'] = pd.to_datetime(df['ts'], utc=True)

# Cutoff for NaN conn_country fallback
cutoff = pd.Timestamp('2024-08-04', tz='UTC')

# Split by group and localize separately
dfs = []

# India
mask_in = df['conn_country'] == 'IN'
df_in = df[mask_in].copy()
df_in['ts_local'] = df_in['ts'].dt.tz_convert('Asia/Kolkata')
dfs.append(df_in)

# US
mask_us = df['conn_country'] == 'US'
df_us = df[mask_us].copy()
df_us['ts_local'] = df_us['ts'].dt.tz_convert('America/New_York')
dfs.append(df_us)

# Qatar
mask_qa = df['conn_country'] == 'QA'
df_qa = df[mask_qa].copy()
df_qa['ts_local'] = df_qa['ts'].dt.tz_convert('Asia/Qatar')
dfs.append(df_qa)

# NaN fallback (US after Aug 4, 2024)
mask_nan_us = df['conn_country'].isna() & (df['ts'] >= cutoff)
df_nan_us = df[mask_nan_us].copy()
df_nan_us['ts_local'] = df_nan_us['ts'].dt.tz_convert('America/New_York')
dfs.append(df_nan_us)

# NaN fallback (India before Aug 4, 2024)
mask_nan_in = df['conn_country'].isna() & (df['ts'] < cutoff)
df_nan_in = df[mask_nan_in].copy()
df_nan_in['ts_local'] = df_nan_in['ts'].dt.tz_convert('Asia/Kolkata')
dfs.append(df_nan_in)
```

```python
# Combine all
df = pd.concat(dfs).sort_index()
```

```python
import re

def clean_platform(value):
    if pd.isna(value):
        return 'Unknown'

    val = str(value).lower()

    if re.search(r'android', val):
        return 'Android'
    elif re.search(r'windows', val):
        return 'Windows'
    elif re.search(r'ios|iphone|ipad|mac|darwin', val):
        return 'iOS'
    elif re.search(r'google cast|chromecast|cast_', val):
        return 'Google Cast'
    else:
        return 'Other'
    
# Apply to platform column
df['platform_clean'] = df['platform'].apply(clean_platform)
```

```python
import pandas as pd
import re

# Step 1: Ensure ts_local is string before applying regex
df_clean['ts_local_clean'] = df_clean['ts_local'].astype(str).str.replace(r'([+-]\d{2}:\d{2})$', '', regex=True)

# Step 2: Convert cleaned strings to datetime (naive, no timezone)
df_clean['ts_local_clean'] = pd.to_datetime(df_clean['ts_local_clean'], errors='coerce')

# Step 3: Extract datetime features
df_clean['Hour'] = df_clean['ts_local_clean'].dt.hour
df_clean['Minute'] = df_clean['ts_local_clean'].dt.minute
df_clean['Second'] = df_clean['ts_local_clean'].dt.second
df_clean['HH:MM:SS'] = df_clean['ts_local_clean'].dt.strftime('%H:%M:%S')

df_clean['Date'] = df_clean['ts_local_clean'].dt.date
df_clean['Day_Name'] = df_clean['ts_local_clean'].dt.day_name()
df_clean['Month'] = df_clean['ts_local_clean'].dt.month
df_clean['Month_Name'] = df_clean['ts_local_clean'].dt.month_name()
df_clean['Year'] = df_clean['ts_local_clean'].dt.year
```
